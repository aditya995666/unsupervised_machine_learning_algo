# **heirarical cluster - unsupervised machine learnig**

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=load_iris()
df.data
```

```python
# feature name
df.feature_names
```

```python
df.target
```

```python
df.target_names
```

```python
df2=pd.DataFrame(df.data,columns=df.feature_names)
df2
```

```python
# split train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df2,df.target,test_size=0.2)
```

```python
# future scalling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# conver four feature to two feature
import sklearn.decomposition as pca
pca=pca.PCA(n_components=2) # componennt 2
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python

x_train_pca.shape
```

```python
x_train_pca
```

```python
plt.scatter(x_train_pca[:,0],x_train_pca[:,1])
plt.xlabel('pca-1')
plt.ylabel('pca-2')
plt.title('2 component PCA')
```

# apply aggromearative cluster

```python
plt.figure(figsize=(15,9))
# to connstruct dendogr
import scipy.cluster.hierarchy as sc
sc.dendrogram(sc.linkage(x_train_pca,method="ward"))
plt.title('dendrogram')
plt.xlabel('sample')
plt.ylabel('distance')
```

```python
# apply agglomerative
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward'   )
cluster.fit(x_train_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(x_train_pca[:,0],x_train_pca[:,1],c=cluster.labels_)
plt.xlabel('pca-1')
plt.ylabel('pca-2')
plt.title('2 component PCA')
```

```python
# silhouette_score
from sklearn.metrics import silhouette_score
sil=[]
for k in range(2,11):
    aglo=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward'   )
    aglo.fit(x_train)
    cluster=silhouette_score(x_train,aglo.labels_)
    sil.append(cluster)
```

```python
plt.plot(range(2,11),sil)
plt.xlabel("n_cluster")
plt.ylabel("silhouette_score")
```
